```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
import os
os.makedirs('landscape_2', exist_ok=True)
```

```python
canvas = np.zeros((256, 256*2, 3), dtype=np.uint8)
for file in os.listdir("landscape"):
    if file == ".DS_Store" or file == "df.csv": continue
    target = plt.imread(f"landscape/{file}")
    X = cv2.resize(target, (128, 128))
    X = cv2.resize(X, (256, 256))
    canvas[:, :256, :] = X
    canvas[:, 256:, :] = target
    plt.axis('off')
    plt.imsave(f"landscape_2/{file}", canvas)
```

```python
plt.imshow(canvas)
```
